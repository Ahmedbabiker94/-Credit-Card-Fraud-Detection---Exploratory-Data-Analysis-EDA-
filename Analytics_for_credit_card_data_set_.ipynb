```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot class distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()
```

shows the distribution of transaction classes, where class 0 represents legitimate transactions and class 1 represents fraudulent transactions. There is a severe class imbalance in the dataset, with legitimate transactions (class 0) numbering nearly 12,000, while fraudulent transactions (class 1) appear to be fewer than 100.

```python
# Plot correlation heatmap
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

The correlation Heat Map shows the linear relationships between variables in the dataset, revealing that most features exhibit very low correlation with each other, suggesting high independence.The diagonal line shows perfect self-correlation, while off-diagonal values are predominantly near zero, with only a few exceptions showing moderate correlation.
Notably, some features such as V12, V14, and V17 exhibit stronger negative correlations with the target variable Class, indicating their potential importance in distinguishing between classes.The original features Time and Amount have minimal correlation with both the transformed features and the class label, suggesting limited predictive value on their own.

```python
#transaction amount distrbution by class
plt.figure(figsize=(10,4))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount Distribution by Class")
plt.xticks([0,1], ['Legit', 'Fraud'])
plt.yscale('log')  # helps with outliers
plt.show()
```

The boxplot compares the distribution of transaction amounts between legitimate and fraudulent transactions. It uses a logarithmic scale on the y-axis to better visualize the wide range of transaction amounts.
Fraudulent transactions tend to involve smaller amounts and show less variation in size compared to legitimate ones, which span a much broader range and include many large-value outliers.


```python
#PCA
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np

# Drop 'Class', 'Time', AND 'AmountBin' columns and handle NaNs
X = df.drop(['Class', 'Time', 'AmountBin'], axis=1, errors='ignore').dropna()  # Drop NaNs and 'AmountBin' here

# Keep the index before converting to NumPy array
X_values = X.to_numpy()

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_values)

plt.figure(figsize=(8,6))

# Get 'Class' values using the original DataFrame index
y = df.loc[X.index, 'Class']
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y, palette='Set1', alpha=0.6)

plt.title("PCA Projection of Transactions")
plt.show()
```

The graph shows projection of transaction data with two classes represented by different colors: red points (Class 0.0, likely non-fraudulent transactions) and blue points (Class 1.0, likely fraudulent transactions). The red points predominantly cluster near the top of the graph around the y-value of 0, spreading horizontally across the x-axis, with the highest density near the origin. In contrast, the blue points form several distinct vertical clusters primarily in the negative y-region, with concentrations around y-values of approximately -10, -20, -40, and even lower. This separation pattern suggests that PCA has effectively differentiated between the two transaction classes, with fraudulent transactions (blue) exhibiting distinct characteristics that cause them to project to lower regions of the 2D space compared to the normal transactions (red), indicating that the underlying features captured by this dimensionality reduction technique successfully highlight differences between fraudulent and legitimate transactions.

```python
df['AmountBin'] = pd.qcut(df['Amount'], q=10, duplicates='drop')
fraud_rate = df.groupby('AmountBin')['Class'].mean()

fraud_rate.plot(kind='barh', figsize=(8,6), title="Fraud Rate by Amount Range")
plt.xlabel("Fraud Rate")
plt.show()
```

The graph shows fraud rates across different transaction amount ranges. Most notably, the smallest transaction range (-0.001, 1.0) has by far the highest fraud rate at approximately 0.027. The next highest fraud rates appear in the largest amount range (149.399, 7712.43) at about 0.005, followed closely by the (1.0, 4.0) range at roughly 0.004. Other amount ranges show minimal fraud rates, with several middle ranges (15.95, 23.772) and (11.85, 15.95) appearing to have nearly zero fraud incidents. This pattern suggests that fraudsters disproportionately target very small transactions (likely hoping they'll go unnoticed) and very large transactions (where the payoff is greatest), while mid-range transactions experience significantly lower fraud rates.
